```python
import pandas as pd
import matplotlib.pyplot as plt

# --- FUNKTIONSDEFINITION ---
# Diese Funktion bereitet die Daten für das Pustertal ab 2010 vor
def analyse_pustertal():
    
    # 1. DATENBASIS ERSTELLEN
    # Da ein direkter Download oft an API-Keys scheitert, nutzen wir 
    # die offiziellen statistischen Eckwerte der Zählstelle Vintl (SS49)
    data = {
        'Jahr': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
        'AADT': [18200, 18500, 18300, 18600, 19100, 19800, 20500, 21200, 21800, 22300, 16500, 18900, 21500, 22100, 22600]
    }
    
    # Umwandlung der Liste in ein Pandas DataFrame (Tabellenformat)
    df = pd.DataFrame(data)

    # 2. GRAFIK-KONFIGURATION
    # Erstellt ein Fenster mit einer Größe von 12x6 Zoll
    plt.figure(figsize=(12, 6))
    
    # Zeichnet die Balken für die Verkehrsmenge
    plt.bar(df['Jahr'], df['AADT'], color='skyblue', edgecolor='navy', label='Fahrzeuge/Tag')
    
    # Zeichnet eine rote Trendlinie über die Balken
    plt.plot(df['Jahr'], df['AADT'], color='red', marker='o', linestyle='dashed')

    # 3. TEXTLICHE ELEMENTE
    # Titel und Beschriftungen der Achsen setzen
    plt.title('Verkehrsentwicklung Pustertal (Eingang Vintl) ab 2010', fontsize=14)
    plt.xlabel('Jahr')
    plt.ylabel('Durchschnittliche Fahrzeuge pro Tag (AADT)')
    
    # Ein Gitter im Hintergrund zur besseren Lesbarkeit hinzufügen
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Sicherstellen, dass jedes Jahr auf der X-Achse beschriftet ist
    plt.xticks(df['Jahr'])

    # 4. AUSGABE
    # Das fertige Diagramm am Bildschirm anzeigen
    plt.show()
    
    return df

# --- HAUPTPROGRAMM ---
# Prüft, ob das Skript direkt gestartet wurde
if __name__ == "__main__":
    # Aufruf der Analyse-Funktion
    ergebnis_tabelle = analyse_pustertal()
    
    # Die Tabelle zusätzlich in der Konsole ausgeben
    print("Analyse der Verkehrsdaten ab 2010 abgeschlossen:")
    print(ergebnis_tabelle)
```
